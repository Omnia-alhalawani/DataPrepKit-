```python
import pandas as pd
class DataPrepKit:
    def read_data (self,file_path):
        file_format=file_path.split('.')[-1].lower()
        if file_format =='csv':
            return pd.read_csv(file_path)
        elif file_format in['xls','xlsx']:
            return pd.read_excel(file_path)
        elif file_format=='json':
            return pd.read_json()
        else:
             raise ValueError("Unsupported file format. Please provide a file with a .csv, .xlsx, .xls, or .json extension.")
           
    def data_summary(self,df):
        print("basic information")
        df.info()
        print("summary")
        print(df.describe())
    def handle_missing_values(self, df, value):
        return df.fillna(value)
    
    def categorical_encoding(self,df):
        categorical_cols = df.select_dtypes(include=['object']).columns
        encoded_df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)  
        return encoded_df
        

    
```

```python

          
   #to use class
#data Reading 
data_prep=DataPrepKit()
file_path=input("enter your path")
df= data_prep.read_data(file_path)
df


     
        

```

```python
#to summarize data
data_prep.data_summary (df)
```

```python
# Handle missing values
filled_data = data_prep.handle_missing_values(df, 0)
filled_data
```

```python
#  Categorical encoding
encoded_data = data_prep.categorical_encoding(df)
encoded_data

```
